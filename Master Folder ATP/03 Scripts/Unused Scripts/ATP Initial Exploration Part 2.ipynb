```python
#drop irrelevant columns

df_matchstats = df_post1991.drop(columns = ['tourney_date',
                                            'draw_size',
                                            'match_num',
                                            'score', 
                                            'winner_entry',
                                            'winner_hand',
                                            'loser_entry',
                                            'loser_hand',
                                            'winner_rank_points',
                                            'loser_rank_points',
                                            'winner_seed',
                                            'winner_ht',
                                            'loser_seed',
                                            'loser_ht',])
```
